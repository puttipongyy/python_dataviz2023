```python
total_vote
```

```python
voted_to_get_partyList = total_vote/100
```

```python
cal_partylist = sum_party
```

```python
cal_partylist['seat'] = sum_party['party_list_vote']/voted_to_get_partyList
```

```python
cal_partylist.reset_index(inplace = True)
```

```python
cal_partylist
```

```python
import math
```

```python
cal_partylist['seat_round'] = [math.floor(i) for i in cal_partylist['seat']]
cal_partylist.head()
```

```python
cal_partylist['seat_decimal'] = cal_partylist['seat'] - cal_partylist['seat_round']
cal_partylist.head()
```

```python
sum_seat_round = cal_partylist['seat_round'].sum()
sum_seat_round
```

```python
top_seat_decimal = cal_partylist.sort_values('seat_decimal', ascending= False).head(100-sum_seat_round)
top_seat_decimal
```

```python
index_list = top_seat_decimal.index
```

```python
index_list
```

```python
sum_up = list()
for i in cal_partylist.index.isin(index_list):
  if i:
    sum_up.append(1)
  else:
    sum_up.append(0)
print(sum_up)
```

```python
cal_partylist['sum_up'] = sum_up
```

```python
cal_partylist
```

```python
cal_partylist['total_seat'] = cal_partylist['seat_round'] + cal_partylist['sum_up']
cal_partylist
```

```python
cal_partylist.info()
```

```python
cal_partylist = cal_partylist.rename(columns={"total_seat" : "party_seat"})
cal_partylist
```
